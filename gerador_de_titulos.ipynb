```python
from openai import OpenAI
import os

```

```python
from google.colab import userdata

# A API KEY deve estar no Secrets do Colab como OPENAI_API_KEY
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
client = OpenAI(api_key=OPENAI_API_KEY)

def perguntar_openai(prompt, model="gpt-4.1-mini"):
    try:
        response = client.chat.completions.create(
            model=model,
            messages=[
                {"role": "user", "content": prompt}
            ],
            max_tokens=500
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"Erro: {e}"
```

```python
def gerar_conteudo_tiktok(texto_video):
    prompt_final = f"""
{PROMPT_BASE}

Texto do vídeo:
\"\"\"{texto_video}\"\"\"
"""
    return perguntar_openai(prompt_final)
```
